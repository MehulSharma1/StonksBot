```python
import yfinance as yf
import icecream
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
stock_code = "MSFT"
ticker = yf.Ticker(stock_code)
```

```python
hist = ticker.history(period="5d",interval="1h")
hist
```

```python
fig = go.Figure(data=[go.Candlestick(x=hist.index,
                open=hist['Open'],
                high=hist['High'],
                low=hist['Low'],
                close=hist['Close'])])
fig.update_layout(xaxis_rangeslider_visible = False, title = f'{stock_code} SHARE PRICE FOR LAST 5 DAYS')

fig.show()
```

```python
fig.write_image("./images/fig1.png")
```
